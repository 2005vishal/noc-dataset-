## So I Have Check The Data Distubution and variance and skewness

```python
q1 = np.percentile(df['Age'], 25)
q3 = np.percentile(df['Age'],75)
iqr = q3 - q1
iqr
upper_limit_age = q3 + 1.5*iqr
lower_limit_age = q1 - 1.5*iqr
print('upper ',upper_limit_age)
print('lower',lower_limit_age)
```

```python
q1 = np.percentile(df['Height'], 25)
q3 = np.percentile(df['Height'],75)
iqr = q3 - q1
iqr
upper_limit_height = q3 + 1.5*iqr
lower_limit_height = q1 - 1.5*iqr
print('upper ',upper_limit_height)
print('lower',lower_limit_height)
```

```python
q1 = np.percentile(df['Weight'], 25)
q3 = np.percentile(df['Weight'],75)
iqr = q3 - q1
iqr
upper_limit_weight = q3 + 1.5*iqr
lower_limit_weight = q1 - 1.5*iqr
print('upper ',upper_limit_weight)
print('lower',lower_limit_weight)
```

```python
q1 = np.percentile(df['Year'], 25)
q3 = np.percentile(df['Year'],75)
iqr = q3 - q1
iqr
upper_limit_year = q3 + 1.5*iqr
lower_limit_year = q1 - 1.5*iqr
print('upper ',upper_limit_year)
print('lower',lower_limit_year)
```

```python
# this new data base create.i have remove all outlier in this dataset
df1 = df[(df['Age'] > lower_limit_age) & (df['Age'] < upper_limit_age) & (df['Height'] > lower_limit_height) & (df['Height'] < upper_limit_height) & (df['Year'] > lower_limit_year) & (df['Weight'] > lower_limit_weight) & (df['Weight'] < upper_limit_weight)] 
```

```python
plt.figure(figsize = (8,20))
plt.subplot(3,1,1)
sns.boxplot(x = df1['Age'])
plt.subplot(3,1,2)
sns.boxplot(x = df1['Height'])
plt.subplot(3,1,3)
sns.boxplot(x = df1['Weight'])
```

```python
def age_c(data):
    if 10<= data <=20:
        return '10-20'
    elif 21<= data <=30:
        return '21-30'
    elif 31<= data <=40:
        return '31 - 40'
df1['Age_Group'] = df1['Age'].apply(age_c)
```
